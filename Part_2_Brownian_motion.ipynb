```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
kxy_vals = np.arange(0.2, 5.1, 0.2)  # Stiffness values (fN/nm)
sigma_xy2 = 1 / kxy_vals              # Variance as a function of stiffness

# Generate numerical data for particles
num_particles = 10000
y_data = []
x_data = []
for kxy in kxy_vals:
    # Simulating positions of particles based on Gaussian distribution
    y_samples = np.random.normal(0, np.sqrt(1/kxy), num_particles)  # corrected variance usage
    x_samples = np.random.normal(0, np.sqrt(1/kxy), num_particles)  # corrected variance usage
    y_data.append(y_samples)
    x_data.append(x_samples)

# Plotting
plt.figure(figsize=(10, 8))

# Subplot for variance
plt.subplot(2, 1, 1)
plt.plot(kxy_vals, sigma_xy2, 'b-', linewidth=2, label='Theoretical (Solid Curve)')
# Adding scatter points: one symbol per stiffness
plt.scatter(kxy_vals, sigma_xy2, color='r', label='Numerical (Symbols)', zorder=3)
plt.xlabel('Stiffness k_{xy} (fN/nm)')
plt.ylabel('Variance _{xy} (nm)')
plt.title('Variance as a function of Stiffness')
plt.legend()
plt.grid(True)

# Panel 2: Position distributions in the y-plane
plt.subplot(2, 1, 2)
y_range = np.arange(-10, 10, 0.1)  # Range for position y
for kxy in [0.2, 1.0, 5.0]:
    # Probability distribution function (assuming Gaussian)
    p_dist = np.exp(-0.5 * (y_range**2) * kxy) / np.sqrt(2 * np.pi * kxy)
    plt.plot(y_range, p_dist / max(p_dist), label=f'k_{kxy} = {kxy:.1f} fN/nm')

plt.xlabel('Position y (nm)')
plt.ylabel('Probability Density')
plt.title('Position Distributions in the y-plane for Different Stiffness')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Scatter plots for y-x plane for three different kxy
for kxy in [0.2, 1.0, 5.0]:
    plt.figure(figsize=(6, 6))
    index = (np.abs(kxy_vals - kxy)).argmin()  # Getting the index of the closest value
    plt.scatter(x_data[index], y_data[index], s=1, alpha=0.5)
    plt.title(f'Scatter Plot in y-x Plane for k_{kxy} = {kxy} fN/nm')
    plt.xlabel('Position x (nm)')
    plt.ylabel('Position y (nm)')
    plt.axhline(0, color='grey', lw=0.5, ls='--')
    plt.axvline(0, color='grey', lw=0.5, ls='--')
    plt.grid(True)
    plt.axis('equal')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
R = 1e-6  # Example value for a parameter (you should define this)
eta = 0.001  # Example value for friction coefficient (you should define this)
time_ = np.linspace(400, 1000, 2000)  # Time range specifically for MSD
stiffness_vals = [0.2, 1.0, 5.0]  # Different stiffness values (arbitrary units)

# Autocorrelation function and Mean Square Displacement function
def position_autocorrelation(t, k, gamma):
    return np.exp(-k * np.abs(t) / gamma)

def mean_square_displacement(t, k, gamma):
    # As time increases, the MSD will eventually plateau due to confinement
    plateau_value = (k * 10) ** 2  # Change this as needed for your plateau level
    return plateau_value * (1 - np.exp(-k * (t - 400) / gamma))

# Plotting Fig. 5 (a) and Fig. 5 (b)
plt.figure(figsize=(7, 7))

# Subplot (a): Position Autocorrelation Function
plt.subplot(2, 1, 1)
time = np.linspace(-10, 10, 200)  # Time range for autocorrelation
for k in stiffness_vals:
    ACF = position_autocorrelation(time, k, 1)  # assuming gamma = 1 for simplicity
    plt.plot(time, ACF, label=f'Stiffness = {k} fN/nm.')

plt.xlabel('Time (t) [ms]')
plt.ylabel('Autocorrelation Function (C(t)) [n.u.]')
plt.title('Position Autocorrelation Function of a Trapped Particle')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.grid(True)

# Subplot (b): Mean Square Displacement
plt.subplot(2, 1, 2)
for k in stiffness_vals:
    MSD = mean_square_displacement(time_, k, 1)  # assuming gamma = 1 for simplicity
    plt.plot(time_, MSD, label=f'Stiffness = {k} fN/nm.')

# Adjusting the plateau time for proper display
gamma = 6 * np.pi * R * eta  # friction coefficient

# Calculate plateau times for visualization (inversely proportional to stiffness)
plateau_times = [gamma / k for k in stiffness_vals]  # Now it's correctly computed
for pt in plateau_times:
    plt.axvline(x=pt, linestyle='--', color='grey', lw=0.7)

plt.xlabel('Time (t) [ms]')
plt.ylabel('Mean Square Displacement (MSD) [nm]')
plt.title('Mean Square Displacement of a Trapped Particle')
plt.xlim(400, 410)
plt.legend()
plt.grid(True)

# Update legends to clarify relationship
plt.legend(title="Effect of Stiffness", loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import correlate

# Parameters
R = 1  # Example parameter
eta = 1  # Example friction coefficient
Dt = 1  # Time step
num_points = 1000  # Number of data points
stiffness_vals = [0.2, 1.0, 5.0]  # Stiffness values

# Generate synthetic position data as a list of arrays
np.random.seed(42)
x = [np.cumsum(np.random.normal(0, 1, num_points)) for _ in stiffness_vals]

# Function to calculate the position autocorrelation function
def position_acf(data, Dt):
    n = len(data)
    r = correlate(data - np.mean(data), data - np.mean(data), mode='full')
    r = r[n-1:]  # Keep only the positive lags
    r /= np.arange(n, 0, -1)  # Normalize correctly
    time_lags = Dt * np.arange(len(r))  # Time lags
    return r, time_lags

# Function to calculate Mean Square Displacement (MSD)
def MSD(data, Dt):
    n = len(data)
    msd = np.zeros(n)

    for tau in range(n):
        if tau < n:
            msd[tau] = np.mean((data[tau:] - data[:-tau])**2) if tau > 0 else np.mean(data**2)
        else:
            msd[tau] = 0

    time_lags = Dt * np.arange(n)  # Time lags
    return msd, time_lags

# Plotting the graphs
plt.figure(figsize=(12, 10))

# Subplot (a): Position Autocorrelation Function
plt.subplot(2, 1, 1)
for idx, stiffness in enumerate(stiffness_vals):
    ACF, time_lags_acf = position_acf(x[idx], Dt)  # Use individual stiffness data
    plt.plot(time_lags_acf, ACF, label=f'Stiffness = {stiffness} n.u.')

plt.xlabel('Time (t) [ms]')
plt.ylabel('Position Autocorrelation Function (C(t)) [n.u.]')
plt.title('Position Autocorrelation Function of a Trapped Particle')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.grid(True)

# Subplot (b): Mean Square Displacement
plt.subplot(2, 1, 2)
for idx, stiffness in enumerate(stiffness_vals):
    msd_values, time_lags_msd = MSD(x[idx], Dt)  # Use individual stiffness data
    plt.plot(time_lags_msd, msd_values, label=f'Stiffness = {stiffness} fN/nm.')

plt.xlabel('Time (t) [ms]')
plt.ylabel('Mean Square Displacement (MSD) [nm]')
plt.title('Mean Square Displacement of a Trapped Particle')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
